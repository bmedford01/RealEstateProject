```python
#import dependencies
import pandas as pd
import datetime as dt
from time import strptime
import re
```

```python
# CSV files to load
csv_hotness = "Market_Hotness.csv"
```

```python
# Read in Files and store into Pandas data frames
df_hotness = pd.read_csv(csv_hotness)
df_hotness.head()
```

```python
df_hotness.tail()
```

```python
# Print all column names for reference
for col in df_hotness.columns:
    print(col)
```

```python
# Separate Month and Year into separate columns/values 
df_hotness['month_date_yyyymm'] = df_hotness['month_date_yyyymm'].apply(str)
df_hotness['Year'] = df_hotness.month_date_yyyymm.str.slice(stop=4)
df_hotness['Month'] = df_hotness.month_date_yyyymm.str.slice(start=4)
```

```python
df_hotness['Date'] = df_hotness['Month'].map(str) + df_hotness['Year'].map(str)
df_hotness.head()
```

```python
df_hotness['Date'] =  pd.to_datetime(df_hotness['Date'], format='%m%Y')
df_hotness.head()
```

```python
# Update Month from numeric to descriptive text
df_hotness['Month'] = df_hotness['Month'].str.replace('12', 'December')
df_hotness['Month'] = df_hotness['Month'].str.replace('11', 'November')
df_hotness['Month'] = df_hotness['Month'].str.replace('10', 'October')
df_hotness['Month'] = df_hotness['Month'].str.replace('01', 'January')
df_hotness['Month'] = df_hotness['Month'].str.replace('02', 'February')
df_hotness['Month'] = df_hotness['Month'].str.replace('03', 'March')
df_hotness['Month'] = df_hotness['Month'].str.replace('04', 'April')
df_hotness['Month'] = df_hotness['Month'].str.replace('05', 'May')
df_hotness['Month'] = df_hotness['Month'].str.replace('06', 'June')
df_hotness['Month'] = df_hotness['Month'].str.replace('07', 'July')
df_hotness['Month'] = df_hotness['Month'].str.replace('08', 'August')
df_hotness['Month'] = df_hotness['Month'].str.replace('09', 'September')
```

```python
# Format price columns with dollar signs, commas and decimal point
df_hotness["median_listing_price"] = df_hotness["median_listing_price"].map("${:,.2f}".format)
```

```python
# Split City and State into separate columns 
df_hotness[['City', 'State']] = df_hotness.zip_name.str.split(",",expand=True)
```

```python
# Update format for scores to two decimal points 
df_hotness["hotness_score"] = df_hotness["hotness_score"].map("{:.2f}".format)
df_hotness["supply_score"] = df_hotness["supply_score"].map("{:.2f}".format)
df_hotness["demand_score"] = df_hotness["demand_score"].map("{:.2f}".format)
df_hotness["ldp_unique_viewers_per_property_vs_us"] = df_hotness["ldp_unique_viewers_per_property_vs_us"].map("{:.2f}".format)
df_hotness["median_listing_price_vs_us"] = df_hotness["median_listing_price_vs_us"].map("{:.2f}".format)
```

```python
df_hotness.head()
```

```python
# Create new dataframe to include columns of interest from original Dataframe.
df_hotness_clean = df_hotness[["month_date_yyyymm",
                         "Date",
                         "Year",
                         "Month",
                         "postal_code",
                         "City",
                         "State",
                         "median_listing_price", 
                         "nielsen_hh_rank",
                         "hotness_rank", 
                         "hotness_score", 
                         "supply_score", 
                         "demand_score",
                         "median_days_on_market",
                         "median_dom_vs_us",
                         "ldp_unique_viewers_per_property_vs_us",
                         "median_listing_price_vs_us"]]

df_hotness_clean.head()
```

```python
# Rename column Names
df_hotness_renamed = df_hotness_clean.rename(columns={"month_date_yyyymm": "Year/Month", 
                                                "postal_code": "Zip Code", 
                                                "median_listing_price": "Median Listing Price",
                                                "nielsen_hh_rank": "Household Count Rank",
                                                "hotness_rank": "Hotness Rank",
                                                "hotness_score": "Hotness Score",
                                                "supply_score": "Supply Score",
                                                "demand_score": "Demand Score",
                                                "median_days_on_market": "Median Days on Market (DOM)",
                                                "median_dom_vs_us": "Median DOM vs. US",
                                                "ldp_unique_viewers_per_property_vs_us": "Viewers per Property vs. US",
                                                "median_listing_price_vs_us": "Median List Price vs. US"})
df_hotness_renamed.head()

```

```python
# Export file as a CSV < without the pandas index, but with the header
df_hotness_renamed.to_csv("Cleaned_Data/MarketHotness_cleaned.csv", index=False, header=True)
```
