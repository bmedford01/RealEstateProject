```python
#import dependencies
import pandas as pd
import datetime as dt
from time import strptime
import re
```

```python
# CSV files to load
csv_2016 = "Historical_2016_Data.csv"
```

```python
# Read in Files and store into Pandas data frames
df_2016 = pd.read_csv(csv_2016)
```

```python
# Separate Month and Year into separate columns/values 
df_2016['month_date_yyyymm'] = df_2016['month_date_yyyymm'].apply(str)
df_2016['Year'] = df_2016.month_date_yyyymm.str.slice(stop=4)
df_2016['Month'] = df_2016.month_date_yyyymm.str.slice(start=4)
```

```python
# Update Month from numeric to descriptive text
df_2016['Month'] = df_2016['Month'].str.replace('12', 'December')
df_2016['Month'] = df_2016['Month'].str.replace('11', 'November')
df_2016['Month'] = df_2016['Month'].str.replace('10', 'October')
df_2016['Month'] = df_2016['Month'].str.replace('01', 'January')
df_2016['Month'] = df_2016['Month'].str.replace('02', 'February')
df_2016['Month'] = df_2016['Month'].str.replace('03', 'March')
df_2016['Month'] = df_2016['Month'].str.replace('04', 'April')
df_2016['Month'] = df_2016['Month'].str.replace('05', 'May')
df_2016['Month'] = df_2016['Month'].str.replace('06', 'June')
df_2016['Month'] = df_2016['Month'].str.replace('07', 'July')
df_2016['Month'] = df_2016['Month'].str.replace('08', 'August')
df_2016['Month'] = df_2016['Month'].str.replace('09', 'September')
```

```python
# Format price columns with dollar signs, commas and decimal point
df_2016["median_listing_price"] = df_2016["median_listing_price"].map("${:,.2f}".format)
df_2016["average_listing_price"] = df_2016["average_listing_price"].map("${:,.2f}".format)
df_2016["median_listing_price_per_square_foot"] = df_2016["median_listing_price_per_square_foot"].map("${:,.2f}".format)
```

```python
# Format Median Square Feet column to include comma for thousands separator 
df_2016["median_square_feet"] = df_2016["median_square_feet"].map("{:,}".format)
```

```python
# Split City and State into separate columns 
df_2016[['City', 'State']] = df_2016.zip_name.str.split(",",expand=True)
```

```python
# Update respective columns to percentage values 
df_2016["median_listing_price_mm"] = 100 * df_2016["median_listing_price_mm"]
df_2016["median_listing_price_mm"] = df_2016["median_listing_price_mm"].map("{:.2f}%".format)
```

```python
df_2016_clean = df_2016[["month_date_yyyymm",
                         "Year",
                         "Month",
                         "postal_code",
                         "City",
                         "State",
                         "median_listing_price", 
                         "average_listing_price",
                         "active_listing_count", 
                         "median_days_on_market", 
                         "new_listing_count", 
                         "price_increased_count",
                         "price_reduced_count",
                         "pending_listing_count",
                         "median_listing_price_per_square_foot",
                         "median_square_feet",
                         "total_listing_count",
                         "pending_ratio"]]
df_2016_clean.head()
```

```python
df_2016_renamed = df_2016_clean.rename(columns={"month_date_yyyymm": "Year/Month", 
                                                "postal_code": "Zip Code", 
                                                "median_listing_price": "Median Listing Price",
                                                "average_listing_price": "Average Listing Price",
                                                "active_listing_count": "Active Listing Count", 
                                                "median_days_on_market": "Median Days on Market",
                                                "new_listing_count": "New Listing Count",
                                                "price_increased_count": "Price Increased Count",
                                                "price_reduced_count": "Price Reduced Count",
                                                "pending_listing_count": "Pending Listing Count",
                                                "median_listing_price_per_square_foot": "Median Price per Square Foot (List Price)",
                                                "median_square_feet": "Median Square Feet",
                                                "total_listing_count": "Total Listing Count",
                                                "pending_ratio": "Pending Ratio"})
df_2016_renamed.head()
```

```python
# Export file as a CSV < without the pandas index, but with the header
df_2016_renamed.to_csv("Cleaned_Data/historical_2016_cleaned.csv", index=False, header=True)
```
