```python
#import dependencies
import pandas as pd
import datetime as dt
from time import strptime
```

```python
# CSV files to load
csv_home_sales = "Home_Sales_Data.csv"
```

```python
# Read in Files and store into Pandas data frames
df_home_sales = pd.read_csv(csv_home_sales)
df_home_sales.head()
```

```python
df_home_sales.tail()
```

```python
# remove data points with missing values
df_home_sales = df_home_sales.dropna()
df_home_sales
```

```python
# Convert Period Begin and Period End column values to datetime format
df_home_sales["period_begin"] = pd.to_datetime(df_home_sales["period_begin"])
df_home_sales["period_end"] = pd.to_datetime(df_home_sales["period_end"])
```

```python
# Split County and State into separate columns 
df_home_sales[['County', 'State']] = df_home_sales.region_name.str.split(",",expand=True)
```

```python
# Sort the dataframe by the period begin column
df_home_sales = df_home_sales.sort_values(by="period_begin")
df_home_sales.head()
```

```python
# Split the Period Begin column into Year, Month and Day columns to retrieve Year and Month values on their own.
df_home_sales['period_begin'] = df_home_sales['period_begin'].apply(str)
df_home_sales[['Year', 'Month', 'Day']] = df_home_sales.period_begin.str.split("-",expand=True)
```

```python
# Update Month from numeric to descriptive text
df_home_sales['Month'] = df_home_sales['Month'].str.replace('12', 'December')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('11', 'November')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('10', 'October')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('01', 'January')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('02', 'February')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('03', 'March')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('04', 'April')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('05', 'May')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('06', 'June')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('07', 'July')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('08', 'August')
df_home_sales['Month'] = df_home_sales['Month'].str.replace('09', 'September')
```

```python
# Update respective columns to percentage values 

df_home_sales["percent_homes_sold_with_price_drops"] = 100 * df_home_sales["percent_homes_sold_with_price_drops"]
df_home_sales["percent_homes_sold_with_price_drops"] = df_home_sales["percent_homes_sold_with_price_drops"].map("{:.2f}%".format)

df_home_sales["percent_active_listings_with_price_drops"] = 100 * df_home_sales["percent_active_listings_with_price_drops"]
df_home_sales["percent_active_listings_with_price_drops"] = df_home_sales["percent_active_listings_with_price_drops"].map("{:.2f}%".format)

df_home_sales["percent_total_price_drops_of_inventory"] = 100 * df_home_sales["percent_total_price_drops_of_inventory"]
df_home_sales["percent_total_price_drops_of_inventory"] = df_home_sales["percent_total_price_drops_of_inventory"].map("{:.2f}%".format)

df_home_sales["percent_homes_sold_above_list"] = 100 * df_home_sales["percent_homes_sold_above_list"]
df_home_sales["percent_homes_sold_above_list"] = df_home_sales["percent_homes_sold_above_list"].map("{:.2f}%".format)
```

```python
# Update respective column values to include two decimal points
df_home_sales["median_new_listing_ppsf"] = df_home_sales["median_new_listing_ppsf"].map("{:.2f}".format)
df_home_sales["median_active_list_ppsf"] = df_home_sales["median_active_list_ppsf"].map("{:.2f}".format)
df_home_sales["avg_offer_to_list"] = df_home_sales["avg_offer_to_list"].map("{:.2f}".format)
df_home_sales["average_sale_to_list_ratio"] = df_home_sales["average_sale_to_list_ratio"].map("{:.2f}".format)
df_home_sales["months_of_supply"] = df_home_sales["months_of_supply"].map("{:.2f}".format)
```

```python
# Format price columns with dollar signs, commas and decimal point

df_home_sales["median_sale_price"] = df_home_sales["median_sale_price"].map("${:,.2f}".format)

df_home_sales["median_active_list_price"] = df_home_sales["median_active_list_price"].map("${:,.2f}".format)
```

```python
# Format price columns with dollar signs and decimal point
# Need to convert these columns from object to float

df_home_sales["median_new_listing_ppsf"] = df_home_sales["median_new_listing_ppsf"].astype(float, errors='raise')
df_home_sales["median_new_listing_ppsf"] = df_home_sales["median_new_listing_ppsf"].map("${:.2f}".format)

df_home_sales["median_active_list_ppsf"] = df_home_sales["median_active_list_ppsf"].astype(float, errors='raise')
df_home_sales["median_active_list_ppsf"] = df_home_sales["median_active_list_ppsf"].map("${:.2f}".format)
```

```python
# filter new dataframe to include only rows that include 1 week for the duration
df_home_sales_filtered = df_home_sales.loc[df_home_sales["duration"] == "1 weeks"]
```

```python
# Print all column names for reference
for col in df_home_sales.columns:
    print(col)
```

```python
df_home_sales_clean = df_home_sales_filtered[[
"County",
"State",
"Year",
"Month",
"period_begin",
"period_end",
"total_homes_sold",
"total_homes_sold_with_price_drops",
"percent_homes_sold_with_price_drops",
"median_sale_price",
"median_sale_ppsf",
"median_days_to_close",
"price_drops",
"percent_active_listings_with_price_drops",
"total_new_listings",
"median_new_listing_price",
"median_new_listing_ppsf",
"inventory",
"total_active_listings",
"age_of_inventory",
"median_active_list_price",
"median_active_list_ppsf",
"avg_offer_to_list",
"average_sale_to_list_ratio",
"median_days_on_market",
"months_of_supply",
"percent_total_price_drops_of_inventory",
"percent_homes_sold_above_list"
]]

df_home_sales_clean.head(25)
```

```python
df_home_sales_renamed = df_home_sales_clean.rename(columns={
    "period_begin": "Period Begin",
    "period_end": "Period End",
    "total_homes_sold": "Homes Sold",
    "total_homes_sold_with_price_drops": "Homes Sold with Price Drop",
    "percent_homes_sold_with_price_drops": "Homes Sold with Price Drop (%)",
    "median_sale_price": "Median Sale Price",
    "median_sale_ppsf": "Median Sales Price per Sqft",
    "median_days_to_close": "Median Days to Close",
    "price_drops": "Price Drops",
    "percent_active_listings_with_price_drops": "Listings with Price Drop (%)",
    "total_new_listings": "New Listings",
    "median_new_listing_price": "Median New List Price",
    "median_new_listing_ppsf": "Median New Listing Price per Sqft",
    "inventory": "Inventory",
    "total_active_listings": "Active Listings",
    "age_of_inventory": "Age of Inventory",
    "median_active_list_price": "Median List Price",
    "median_active_list_ppsf": "Median List Price per Sqft",
    "avg_offer_to_list": "Average Offer to List",
    "average_sale_to_list_ratio": "Average Sale to List Ratio",
    "median_days_on_market": "Median Days on Market",
    "months_of_supply": "Months of Supply",
    "percent_total_price_drops_of_inventory": "Price Drops of Inventory (%)",
    "percent_homes_sold_above_list": "Homes Sold Above List (%)"
})

df_home_sales_renamed.head()
```

```python
# Export file as a CSV < without the pandas index, but with the header
df_home_sales_renamed.to_csv("Cleaned_Data/HomeSalesData_cleaned.csv", index=False, header=True)
```
